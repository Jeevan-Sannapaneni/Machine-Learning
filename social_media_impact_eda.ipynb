```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_media_impact_on_life.csv')
```

```python
df['Gender'].value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'],bins=15,color='skyblue',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Avg_Daily_Usage_Hours'],bins=15,color='orange',edgecolor='black')
plt.title("Daily Social Media Usage (Hours)")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Mental_Health_Score'],bins=15,color='green',edgecolor='black')
plt.title("Mental Health Score Distribution")
plt.show()
```

```python
df['Overall_Impact'].value_counts().plot(kind='bar',color=['green','red','gray'])
plt.title("Overall Impact Distribution")
plt.ylabel("Count")
plt.show()
```

```python
df['Most_Used_Platform'].value_counts().plot(kind='bar')
plt.title("Most Used Social Media Platform")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Avg_Daily_Usage_Hours'],df['Mental_Health_Score'],alpha=0.5)
plt.title("Usage vs Mental Health")
plt.xlabel("Usage Hours")
plt.ylabel("Mental Health Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Avg_Daily_Usage_Hours'],df['Sleep_Hours_Per_Night'],alpha=0.5)
plt.title("Usage vs Sleep")
plt.xlabel("Usage Hours")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Most_Used_Platform',y='Mental_Health_Score',data=df)
plt.xticks(rotation=45)
plt.title("Platform vs Mental Health")
plt.show()
```
